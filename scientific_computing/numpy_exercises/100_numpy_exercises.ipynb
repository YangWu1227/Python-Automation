#### 10. Find indices of non-zero elements from arrays (★☆☆)

````python
import numpy as np

# One dimensional
np.nonzero(a=[1, 2, 0, 0, 4, 0])
````

````python
# Two dimensional
array = np.array([[1, 2, 0], [0, 0, 4], [0, 0, 0]])
array
# Equivalent to array.nonzero()
np.nonzero(a=[[1, 2, 0], [0, 0, 4], [0, 0, 0]])
````

In this example above, a tuple of two arrays was returned-- one for each dimension. The first array represents the row indices where non-zero values are found, and the second array represents the column indices where non-zero values are found. For 2d-arrays, each element of the first array represents the `ith` entry and each element of the second array represents the `jth` entry of non-zero elements. 

````python
# Return the i, j indices of the nonzero elements of 'array'
for entry in list(zip(array.nonzero()[0], array.nonzero()[1])):
    entry
````

We can also subset and print the non-zero elements of `array` using:

````python
array[array.nonzero()]
````

A common use case of `nonzero` is to find the indices of an array, where a condition is True.

````python
# Boolean array resulting from element-wise comparison
array >= 2
# Find the True elements of 'array'
array[np.nonzero(array >= 2)]
````

The x values the first elements of the tuples (x, y) and the y values are the second elements of the tuples (x, y). The x values are filled with `val` vertically while the y values are filled with `val` horizontally. Another useful application of `np.meshgrid` is creating rolling window matrices. Suppose we wish to conduct a rolling window analysis with:

* `window_width` = 20 (20 units of time create one window)

* `window_shift` = 5 (Shift 5 units of time before apply the next rolling window of length `window_width`)

````python
window_width = 20
window_shift = 5
````

Suppose we have 1,000,000 rows of data: 

````python
# Indices
x = np.arange(0, 1000000, 1)
x
````

We can create the rolling window as follow:

* Create an array with `(x.size - 20)` elements with step size `window_shift`, since we only wish to create complete windows with full size `window_width`

* Create another array with `window_width` elements

````python
A, B = np.meshgrid(
    np.arange(start=0, stop=x.size - window_width, step=window_shift),
    np.arange(window_width),
    indexing="xy",  # Cartesian indexing
)
A.shape, B.shape
````

````python
A, B
````

The matrix A is a `(20, 199996)` matrix with row values that are the starting indices for each window. The matrix B is a `(20, 199996)` matrix with columns values that are the within rolling window indices. Taking the sum of the two matrices:

* For each of the 20 rows of starting indices 0, 5, 10, 15, ..., 999975 we progressively added 0, 1, 2, ..., 19:
  
  *  The first row of starting indices 0, 5, 10, 15, ..., 999975 gets incremented by 1
  *  The second row starting indices 0, 5, 10, 15, ..., 999975 gets incremented by 2
  *  and so on... until we added the full `window_width`

````python
(A + B)
````

Then, taking the transpose of the resultant matrix above would yield the rolling window matrix, which is a `num_of_window x window.width` matrix.

````python
rolling_window_matrix = (A + B).T
rolling_window_matrix
````

````python
rolling_window_matrix.shape
````

Given the data size of 1,000,000 rows, there are 199,996 rolling windows each with window width 20. The indices of the rows from the original data of 1,000,000 rows used to created each window are contained in each row of the rolling window matrix.

````python
import numpy as np


def find_closest(array, value):
    """
    Find the closest value in an array to a given value.

    Parameters
    ----------
    array : array_like
        Array of values.
    value : scalar
        Value to find the closest value to.

    Returns
    -------
    closest_value : scalar
        Closest value in array to value.
    """
    array = np.asarray(array)
    idx = (
        np.abs(array - value)
    ).argmin()  # Return the indices of the minimum values along an axis or that of the first element if there are ties
    return array[idx]
````

#### 63. Create an array class that has a name attribute (★★☆)

The class below has a `__array_finalize__()` method. The `self` is a new object resulting from `ndarray.__new__(NamedArray, ...)`, therefore it only has attributes that the `ndarray.__new__ constructor` gave it - those of a standard `ndarray`.

We could have got to the `ndarray.__new__` call in 3 ways:

* From an explicit constructor - e.g. `NamedArray()` when `obj` is not None and is already created by `super().__new__(subtype, shape, dtype, buffer, offset, strides, order)`

* From view casting - e.g `arr.view(InfoArray)` where `obj` is `arr` and so `type(obj)` can be `NamedArray`

* From new-from-template - e.g `obj[:3]` so `type(obj)` is `NamedArray`

---

Note that the line `self.name = getattr(obj, name='name', default=None)` is in `__array_finalize__`, rather than in the `__new__` method, because this method `__array_finalize__` sees all creation of `default` objects - with the `NamedArray.__new__` constructor but also with `arr.view(NamedArray)`.

````python
import numpy as np


class NamedArray(np.ndarray):
    def __new__(cls, array, name="default_name"):
        # Input array is an already formed np.ndarray instance
        # We first cast it to be our class type
        obj = np.asarray(array).view(cls)
        obj.name = name
        # Return the newly created class instance
        return obj

    def __array_finalize__(self, obj):
        if obj is None:
            return
        self.name = getattr(obj, "name", "default_name")
````

The object can be instantiated as follows with an explicit constructor:

````python
obj_exp_constr = NamedArray(np.arange(10))

print("The name of the array is:", obj_exp_constr.name)
print("The array has class:", obj_exp_constr.__class__)
````

#### 66. Considering a (w,h,3) image of (dtype=ubyte), compute the number of unique colors (★★☆)

````python
import numpy as np

# Random image of size 256 x 256 where each pixel is represented by a 3-byte color code
w, h = 256, 256
I = np.random.randint(0, 4, (h, w, 3), dtype=np.uint8)

# View each pixel as a single 24-bit integer, rather than three 8-bit bytes
I24 = np.dot(I.astype(np.uint32), [1, 256, 65536])

# Count unique colors
n = len(np.unique(I24))
print(n)
````

The code converts the image from the original 3-byte representation to a single 24-bit integer per pixel. This is achieved by first converting each element of the `I` array to an unsigned 32-bit integer, and then taking the dot product of the resulting array with the matrix `[1, 256, 65536]`. This creates a single 24-bit integer for each pixel, where the first 8 bits represent the blue color channel, the second 8 bits represent the green channel, and the last 8 bits represent the red channel.

Each color channel in the image is represented by an 8-bit value, which can take on values from 0 to 255. To combine the 3 color channels into a single 24-bit integer value, the blue channel is multiplied by 1 (which is equivalent to shifting it left by 0 bits), the green channel is multiplied by 256 (which is equivalent to shifting it left by 8 bits), and the red channel is multiplied by 65536 (which is equivalent to shifting it left by 16 bits). These products are then added together to produce a single 24-bit integer value for each pixel. So the values in `[1, 256, 65536]` are used as weights to create a weighted sum of the 8-bit color values, in order to obtain a single 24-bit integer value for each pixel.

Finally, the code uses the np.unique function to count the number of unique color values in the image. Since the image contains 4 different color codes, the value of `n` printed by the code will be 4.

#### 78. Consider 2 sets of points P0, P1 describing lines (2d) and a point p, how to compute distance from p to each line i (P0[i], P1[i])? (★★★)

If the line passes through two points $P_1=\left(x_1, y_1\right)$ and $P_2=\left(x_2, y_2\right)$ then the distance of $\left(x_0, y_0\right)$ from the line is given by the following [formula](https://en.wikipedia.org/wiki/Distance_from_a_point_to_a_line#Line_defined_by_two_points):

$$
\operatorname{distance}\left(P_1, P_2,\left(x_0, y_0\right)\right)=\frac{\left|\left(y_2-y_1\right) x_0-\left(x_2-x_1\right) y_0+x_2 y_1-y_2 x_1\right|}{\sqrt{\left(y_2-y_1\right)^2+\left(x_2-x_1\right)^2}}
$$


The denominator of this expression is the distance between $P_1$ and $P_2$. The numerator is twice the area of the triangle with its vertices at the three points, $\left(\mathrm{x}_0, \mathrm{y}_0\right), P_1$ and $P_2$.

````python
import numpy as np

p_1 = np.random.uniform(-10, 10, (10, 2))
p_2 = np.random.uniform(-10, 10, (10, 2))
p = np.random.uniform(-10, 10, (1, 2))

p_1, p_2, p
````

````python
def distance_point_to_lines(
    p: np.ndarray, p_1: np.ndarray, p_2: np.ndarray
) -> np.ndarray:
    # Coordinates of p
    x_0, y_0 = p.T
    x_1, y_1 = p_1.T
    x_2, y_2 = p_2.T

    numerators: np.ndarray = np.abs(
        (y_2 - y_1) * x_0 - (x_2 - x_1) * y_0 + x_2 * y_1 - y_2 * x_1
    )
    denominators: np.ndarray = np.sqrt((y_2 - y_1) ** 2 + (x_2 - x_1) ** 2)
    return numerators / denominators
````

#### 79. Consider 2 sets of points P0, P1 describing lines (2d) and a set of points P, how to compute distance from each point j (P[j]) to each line i (P0[i],P1[i])? (★★★)

This is a good application of broadcasting:

````python
import time

import numpy as np

p_1 = np.random.uniform(-10, 10, (10, 2))
p_2 = np.random.uniform(-10, 10, (10, 2))
p = np.random.uniform(-10, 10, (5, 2))

p_1, p_2, p
````

````python
def distance_points_to_lines(
    p: np.ndarray, p_1: np.ndarray, p_2: np.ndarray
) -> np.ndarray:
    x_0, y_0 = p.T  # Shape -> (n points, )
    x_1, y_1 = p_1.T  # Shape -> (n lines, )
    x_2, y_2 = p_2.T  # Shape -> (n lines, )

    # Displacement vector coordinates from p_1 -> p_2
    dx = x_2 - x_1  # Shape -> (n lines, )
    dy = y_2 - y_1  # Shape -> (n lines, )

    # The 'cross product' term
    cross_term = x_2 * y_1 - y_2 * x_1  # Shape -> (n lines, )

    # Broadcast x_0, y_0 (n points, 1) and dx, dy, cross_term (1, n lines) -> (n points, n lines)
    numerator = np.abs(
        dy[np.newaxis, :] * x_0[:, np.newaxis]
        - dx[np.newaxis, :] * y_0[:, np.newaxis]
        + cross_term[np.newaxis, :]
    )
    denominator = np.sqrt(dx**2 + dy**2)  # Shape -> (n lines, )

    # Shape (n points, n lines) / (1, n_lines) -> (n points, n lines)
    return numerator / denominator[np.newaxis, :]
````

````python
distance_points_to_lines(p, p_1, p_2)
````

#### 80. Consider an arbitrary array, write a function that extract a subpart with a fixed shape and centered on a given element (pad with a `fill` value when necessary) (★★★)

````python
from typing import Tuple

import numpy as np


def extract_centered_window(
    arrary: np.ndarray,
    center_idx: Tuple[int, int],
    window_shape: Tuple[int, int],
    fill_value: int = 0,
):
    """
    Extract a subarray with fixed shape centered on a given element, padding if necessary.

    Parameters:
    -----------
    arrary : np.ndarray
        Input array
    center_idx : Tuple[int, int]
        Index of the center element for each dimension
    window_shape : Tuple[int, int]
        Shape of the window to extract
    fill_value : int
        Value to use for padding (default: 0)

    Returns:
    --------
    numpy.ndarray
        Extracted window with requested shape
    """
    array = np.asarray(arrary)
    center_idx = np.asarray(center_idx)
    window_shape = np.asarray(window_shape)

    # Calculate window extents
    starts = center_idx - window_shape // 2
    ends = starts + window_shape

    # Calculate padding needed
    pad_before = np.maximum(0, -starts)
    pad_after = np.maximum(0, ends - array.shape)
    padding = list(zip(pad_before, pad_after))

    # Pad array if necessary
    if np.any(pad_before > 0) or np.any(pad_after > 0):
        array = np.pad(array, padding, mode="constant", constant_values=fill_value)
        # Adjust starts and ends to account for padding
        starts = starts + pad_before
        ends = ends + pad_before

    # Create slice objects for each dimension
    slices = tuple(slice(start, end) for start, end in zip(starts, ends))

    return array[slices]
````

#### Understanding Start and End Calculations

The **starts** and **ends** in this context define the boundaries of the window for extracting a subarray centered around a specified index (`center_idx`) while considering the window size (`window_shape`). 

The general formula for computing `starts` and `ends` is based on how the window aligns with the array, accounting for edge cases (e.g., when the window extends beyond the array boundaries).

#### General Formulas

#### 1. **Starts**:

$$
\text{starts} = \text{center\_idx} - \left\lfloor \frac{\text{window\_shape}}{2} \right\rfloor
$$

- The **center index** is the middle point of the window in the array.
- Subtracting the floor $\left\lfloor \frac{\text{window\_shape}}{2} \right\rfloor$ ensures that the window is symmetrically centered around `center_idx`.

#### 2. **Ends**:

$$
\text{ends} = \text{starts} + \text{window\_shape}
$$

- Once the starting point (`starts`) is computed, adding the `window_shape` gives the corresponding ending index, ensuring the correct size of the window.

#### Why These Formulas?

- The subtraction of $\left\lfloor \frac{\text{window\_shape}}{2} \right\rfloor$ ensures that the window extends evenly to the left and right (or top and bottom in 2D arrays) of the `center_idx`.

- The addition of `window_shape` ensures that the window captures the correct number of rows and columns as specified by its shape.

#### Adjustments for Out-of-Bounds Windows

When the computed `starts` or `ends` fall outside the array boundaries, **padding** is applied, and the indices are adjusted:

#### 3. **Padding Calculation**:
- **Pad Before**:
  $$
  \text{pad\_before} = \max(0, -\text{starts})
  $$
  - Padding is required before the array if `starts < 0`.

- **Pad After**:
  $$
  \text{pad\_after} = \max(0, \text{ends} - \text{array.shape})
  $$
  - Padding is required after the array if `ends > array.shape`.

#### 4. **Adjusted Starts and Ends**:
Once padding is applied, `starts` and `ends` are shifted by the padding amounts:
$$
\text{adjusted\_starts} = \text{starts} + \text{pad\_before}
$$
$$
\text{adjusted\_ends} = \text{ends} + \text{pad\_before}
$$

#### Why Does This Work?

1. **Starts** anchor the window at the top-left corner (or equivalent in higher dimensions).
2. **Ends** define where the window extends, ensuring the dimensions are correct.
3. The combination of padding and adjustments ensures that even if the window extends outside the array, the operation remains valid by filling missing regions with the `fill_value`.

#### General Case in N Dimensions

For arrays of arbitrary dimensionality (\(N\)-dimensional):
- `starts` and `ends` are calculated independently for each dimension \(i\):
  $$
  \text{starts}[i] = \text{center\_idx}[i] - \left\lfloor \frac{\text{window\_shape}[i]}{2} \right\rfloor
  $$
  $$
  \text{ends}[i] = \text{starts}[i] + \text{window\_shape}[i]
  $$
- Padding and adjustments are also applied per dimension.

````python
# Example 1: 1D array with padding
array_1 = np.arange(10)
center_1 = (7,)
window_1 = (12,)
result_1 = extract_centered_window(array_1, center_1, window_1, fill_value=0)

# Example 2: 2D array with padding
array_2 = np.arange(9).reshape(3, 3)
center_2 = (0, 0)
window_2 = (3, 3)
result_2 = extract_centered_window(array_2, center_2, window_2, fill_value=-1)


examples = {
    "Example 1": {
        "Input": array_1,
        "Center": center_1,
        "Window Shape": window_1,
        "Result": result_1,
    },
    "Example 2": {
        "Input": array_2,
        "Center": center_2,
        "Window Shape": window_2,
        "Result": result_2,
    },
}

for example, data in examples.items():
    print(f"{example}:")
    print(f"Input:\n{data['Input']}")
    print(f"Center: {data['Center']}")
    print(f"Window Shape: {data['Window Shape']}")
    print(f"Result:\n{data['Result']}\n")
````

#### Generalized Formulas for Sliding Window Extraction Using `as_strided`

#### Notation

- **Input Array**  

  Let the input array be an $N$-dimensional NumPy array of shape:
  $$
  (D_0, \, D_1, \, \ldots, \, D_{N-1}),
  $$
  and strides:
  $$
  (\text{stride}_0, \, \text{stride}_1, \, \ldots, \, \text{stride}_{N-1}).
  $$

- **Window Shape**  

  We define a window of shape
  $$
  (w_0, \, w_1, \, \ldots, \, w_{N-1}).
  $$
  We assume $w_i \le D_i$, meaning the window does not exceed the size of any dimension.

#### Output Shape

In the 2D case, the output shape was:
$$
\bigl(D_0 - w_0 + 1, \; D_1 - w_1 + 1,\; w_0,\; w_1\bigr).
$$

##### Generalizing to $ N $ Dimensions

For each dimension $i$:

1. The number of **sliding positions** along the $i$-th dimension is
   $$
   D_i - w_i + 1.
   $$
   - This is how many distinct windows we can place along dimension $i$.

2. The **window content** size along the $i$-th dimension is simply $w_i$.

Therefore, the full output shape in $N$ dimensions has two parts:

1. **Sliding Positions** (first $N$ axes):  
   $$
   \left(D_0 - w_0 + 1,\; D_1 - w_1 + 1,\; \ldots,\; D_{N-1} - w_{N-1} + 1 \right)
   $$

2. **Window Size** (last $N$ axes):  
   $$
   (w_0,\; w_1,\; \ldots,\; w_{N-1})
   $$

Putting these together, the output shape is a $2N$-dimensional array:

$$
\text{Output Shape} = 
\Bigl(D_0 - w_0 + 1,\; D_1 - w_1 + 1,\; \ldots,\; D_{N-1} - w_{N-1} + 1,\; w_0,\; w_1,\; \ldots,\; w_{N-1}\Bigr).
$$

#### Strides

Recall the 2D case:  
$$
(\text{stride}_0,\; \text{stride}_1,\; \text{stride}_0,\; \text{stride}_1).
$$

##### Why Are the Strides Repeated?

- **First $N$ strides** (the “sliding” part): moving from one window to the **next** window along dimension $i$ simply advances by 1 element in that dimension in the original array, which is $\text{stride}_i$.

- **Last $N$ strides** (the “within-window” part): moving within the same window (i.e., stepping along dimension $i$ *inside* each extracted window) also advances by 1 element along dimension $i$ in the original array, hence $\text{stride}_i$ again.

##### General $ N $-Dimensional Strides

Following the same logic, each dimension contributes two strides:

1. The stride used to move **between** windows along dimension $i$.
2. The stride used to move **within** a window along dimension $i$.

Hence, for dimension $i$, both of these are just $\text{stride}_i$. Thus, the full stride tuple for the output array is of length $2N$:

$$
\text{Output Strides} = 
\Bigl(\text{stride}_0, \; \text{stride}_1, \; \ldots, \; \text{stride}_{N-1}, \;
      \text{stride}_0, \; \text{stride}_1, \; \ldots, \; \text{stride}_{N-1}\Bigr).
$$

##### Skipping Steps (Generalized Striding)

If we want to **skip steps** along any dimension—say, only take every $k$-th element—we can *adjust* these strides accordingly. Instead of advancing by $\text{stride}_i$ every time, we can advance by $k \times \text{stride}_i$. 

1. **Moving Between Windows**:
   - If we want to jump every $k$ elements along dimension $i$ when moving to the *next* window, we would replace the **first** $\text{stride}_i$ for that dimension with $k \cdot \text{stride}_i$.

2. **Moving Within the Window**:
   - If we want each “internal step” within the window to jump by $m$ elements along dimension $i$, we would replace the **second** $\text{stride}_i$ for that dimension with $m \cdot \text{stride}_i$.

Thus, if we had an original stride for dimension $i$ of $\text{stride}_i$, we could define the two strides for dimension $i$ as:
$$
(\;k \times \text{stride}_i,\; m \times \text{stride}_i\;)
$$
where:
- $k$ is the step size *between* windows,
- $m$ is the step size *within* each window.

In practice, we would set these values depending on how we want to sample or skip elements.

#### Summary of the General Formulas

##### **Output Shape**

$$
\underbrace{\bigl(D_0 - w_0 + 1,\; D_1 - w_1 + 1,\; \ldots,\; D_{N-1} - w_{N-1} + 1\bigr)}_{\text{sliding positions}} \;\times\;
\underbrace{\bigl(w_0,\; w_1,\; \ldots,\; w_{N-1}\bigr)}_{\text{window shape}}.
$$

In total, this is a $2N$-dimensional shape: the first $N$ axes correspond to the number of windows in each dimension, and the last $N$ axes correspond to the window content.

##### **Output Strides**

$$
\underbrace{\bigl(\text{stride}_0,\; \text{stride}_1,\; \ldots,\; \text{stride}_{N-1}\bigr)}_{\text{sliding between windows}} \;\times\;
\underbrace{\bigl(\text{stride}_0,\; \text{stride}_1,\; \ldots,\; \text{stride}_{N-1}\bigr)}_{\text{stepping inside each window}}.
$$

````python
window_size = (2, 2, 2)
x = np.arange(5 * 5 * 5).reshape((5, 5, 5)).astype(np.int32)
x
````

````python
(stride_0, stride_1, stride_2) = x.strides
(stride_0, stride_1, stride_2)
````

#### Example in 3 Dimensions

Suppose:

- Input array shape: $(D_0, D_1, D_2) = (5, 6, 7)$  
- Window shape: $(w_0, w_1, w_2) = (3, 2, 4)$  
- Input array strides: $\text{stride}_0, \text{stride}_1, \text{stride}_2$

1. **Output Shape**:
   $$
   \begin{aligned}
   &\bigl(D_0 - w_0 + 1,\; D_1 - w_1 + 1,\; D_2 - w_2 + 1,\; w_0,\; w_1,\; w_2\bigr) \\
   &= \bigl(5 - 3 + 1,\; 6 - 2 + 1,\; 7 - 4 + 1,\; 3,\; 2,\; 4\bigr) \\
   &= (3, \; 5, \; 4,\; 3, \; 2,\; 4).
   \end{aligned}
   $$

2. **Output Strides**:
   $$
   \bigl(\text{stride}_0,\; \text{stride}_1,\; \text{stride}_2,\; \text{stride}_0,\; \text{stride}_1,\; \text{stride}_2\bigr).
   $$

By passing these shape and stride parameters to `numpy.lib.stride_tricks.as_strided`, we get an array of shape $(3, 5, 4, 3, 2, 4)$ where each “slice” along the first 3 dimensions corresponds to a sliding window of shape $(3,2,4)$ in the original data.

````python
x = np.arange(8 * 7 * 6 * 5).reshape(8, 7, 6, 5).astype(np.int32)

# Strides in each dimension
stride_0, stride_1, stride_2, stride_3 = x.strides

# Window shape
w_0, w_1, w_2, w_3 = 3, 3, 2, 2

# Number of windows in each dimension
D_0, D_1, D_2, D_3 = x.shape
num_0 = D_0 - w_0 + 1
num_1 = D_1 - w_1 + 1
num_2 = D_2 - w_2 + 1
num_3 = D_3 - w_3 + 1

# Output shape = (num_0, num_1, num_2, num_3, w_0, w_1, w_2, w_3)
output_shape = (num_0, num_1, num_2, num_3, w_0, w_1, w_2, w_3)

# Skips: (k_0, k_1, k_2, k_3) and (m_0, m_1, m_2, m_3)
k_0, k_1, k_2, k_3 = 2, 1, 2, 1
m_0, m_1, m_2, m_3 = 1, 2, 2, 3

# Construct output strides
output_strides = (
    k_0 * stride_0,
    k_1 * stride_1,
    k_2 * stride_2,
    k_3 * stride_3,
    m_0 * stride_0,
    m_1 * stride_1,
    m_2 * stride_2,
    m_3 * stride_3,
)

sliding_view = np.lib.stride_tricks.as_strided(
    x, shape=output_shape, strides=output_strides
)

print("Output Shape =", sliding_view.shape)
````

#### Example in 4 Dimensions (With Skips Along Each Dimension)

Suppose:

- **Input array shape**:  
  $$(D_0, D_1, D_2, D_3) = (8, 7, 6, 5)$$

- **Window shape**:  
  $$(w_0, w_1, w_2, w_3) = (3, 3, 2, 2)$$

- **Input array strides**:  
  $$(\text{stride}_0, \text{stride}_1, \text{stride}_2, \text{stride}_3)$$  
  Each $\text{stride}_i$ is the number of **bytes** to move when we increase the index by 1 along dimension \(i\).

- **Skips** (step sizes):  

  For each dimension \(i\), let:
  - $k_i$: the step size **between windows** along dimension $i$
  - $m_i$: the step size **within a single window** along dimension $i$

  Suppose:
  $$
  (k_0,\, k_1,\, k_2,\, k_3) = (2,\, 1,\, 2,\, 1)
  $$
  $$
  (m_0,\, m_1,\, m_2,\, m_3) = (1,\, 2,\, 2,\, 3)
  $$
  For example, along dimension 0, we skip by 2 elements when moving to the next window, but inside each window, we move by 1 element at a time.

##### Output Shape

We first compute the **number of windows** along each dimension:
$$
D_i - w_i + 1.
$$

- Dimension 0: $8 - 3 + 1 = 6$
- Dimension 1: $7 - 3 + 1 = 5$
- Dimension 2: $6 - 2 + 1 = 5$
- Dimension 3: $5 - 2 + 1 = 4$

So, the **first 4 dimensions** of the output represent how many windows we can slide over:
$$
(6,\, 5,\, 5,\, 4).
$$

The **last 4 dimensions** represent each window’s shape:
$$
(w_0, w_1, w_2, w_3) = (3,\, 3,\, 2,\, 2).
$$

Hence, the **output shape** is:
$$
\bigl(6,\, 5,\, 5,\, 4,\; 3,\, 3,\, 2,\, 2\bigr).
$$


##### Output Strides

Normally (with no skipping), we would just repeat each $\text{stride}_i$ **twice**:

$$
(\text{stride}_0,\, \text{stride}_1,\, \text{stride}_2,\, \text{stride}_3,\;
 \text{stride}_0,\, \text{stride}_1,\, \text{stride}_2,\, \text{stride}_3).
$$

But here, each dimension $i$ has two distinct stride factors:

1. **Between windows**: $k_i \times \text{stride}_i$
2. **Within each window**: $m_i \times \text{stride}_i$

Given our chosen $k_i$ and $m_i$:

- Dimension 0: between = $2\,\text{stride}_0$, within = $1\,\text{stride}_0$
- Dimension 1: between = $1\,\text{stride}_1$, within = $2\,\text{stride}_1$
- Dimension 2: between = $2\,\text{stride}_2$, within = $2\,\text{stride}_2$
- Dimension 3: between = $1\,\text{stride}_3$, within = $3\,\text{stride}_3$

So, the **output strides** (length \(2 \times 4 = 8\)) become:

$$
\bigl(
2\,\text{stride}_0,\; 1\,\text{stride}_1,\; 2\,\text{stride}_2,\; 1\,\text{stride}_3,\;
1\,\text{stride}_0,\; 2\,\text{stride}_1,\; 2\,\text{stride}_2,\; 3\,\text{stride}_3
\bigr).
$$

- **First 4 values** = strides to advance **to the next window** in dimensions 0, 1, 2, 3.
- **Last 4 values** = strides to advance **inside** each window in dimensions 0, 1, 2, 3.

#### Important Caveats

1. **Memory Safety**  
   Using `as_strided` can be **dangerous** if the resulting view extends beyond the allocated memory of the array. Always ensure that the window extraction does not lead to out-of-bounds memory access.

2. **Contiguity**  
   If the original array is not C-contiguous (or has other forms of memory lawet, such as Fortran order), the strides array will differ. Always read the original `array.strides` property for the correct dimension ordering.

3. **Overlapping Data**  
   The returned view will have overlapping memory regions across different window positions. Be aware of this if we perform in-place modifications.

#### 85. Create a 2D array subclass such that Z[i, j] == Z[j, i] (★★★)

````python
from __future__ import annotations

from typing import Any, Optional, Tuple, TypeVar, Union, overload

import numpy as np
import numpy.typing as npt

# Type variable for the class
T = TypeVar("T", bound="SymmetricArray")

# Type aliases for clarity
ArrayLike = Union[np.ndarray, list[list[float]], list[list[int]]]
IndexType = Union[Tuple[int, int], int]
DType = TypeVar("DType", bound=np.generic)


class SymmetricArray(np.ndarray[Any, np.dtype[DType]]):
    """
    A 2D numpy array subclass that maintains symmetry across its diagonal.

    This class ensures that for any indices i,j: Z[i,j] == Z[j,i].
    The symmetry is maintained during initialization and assignment operations.

    Parameters
    ----------
    input_array : array_like
        Input array to create the symmetric array from. Must be 2D and square.
    dtype : numpy.dtype, optional
        The desired data type of the array. If not specified, the type will be
        determined from the input array.

    Returns
    -------
    SymmetricArray
        A new instance of SymmetricArray with symmetric elements.

    Raises
    ------
    ValueError
        If the input array is not 2D or not square.

    Examples
    --------
    >>> arr = np.array([[1, 2], [3, 4]])
    >>> sym = SymmetricArray(arr)
    >>> sym[0, 1] = 5
    >>> sym[1, 0]  # Returns 5 due to symmetry
    5
    """

    @overload
    def __new__(cls: type[T], input_array: ArrayLike, dtype: None = None) -> T: ...

    @overload
    def __new__(
        cls: type[T], input_array: ArrayLike, dtype: np.dtype[DType]
    ) -> SymmetricArray[DType]: ...

    def __new__(
        cls: type[T], input_array: ArrayLike, dtype: Optional[np.dtype[Any]] = None
    ) -> Union[T, SymmetricArray[Any]]:
        # Convert input array to numpy array if it isn't already
        array: np.ndarray = np.asarray(input_array, dtype=dtype)

        # Check if the input array is 2D
        if array.ndim != 2:
            raise ValueError("Input array must be 2D")

        # Check if the input array is square
        if array.shape[0] != array.shape[1]:
            raise ValueError("Input array must be square")

        # Create symmetric array from input
        # Take average of array and its transpose to ensure symmetry
        symmetric: np.ndarray = (array + array.T) / 2

        # Create the ndarray instance
        obj = symmetric.view(cls)
        return obj

    def __array_finalize__(self, obj: Optional[np.ndarray]) -> None:
        """
        Handle array creation by view casting and new-from-template.

        Parameters
        ----------
        obj : numpy.ndarray or None
            The parent array from which this array is created.
            None if the array is being created from scratch.
        """
        if obj is None:
            return

    @overload
    def __setitem__(self, index: int, value: Union[float, int]) -> None: ...

    @overload
    def __setitem__(self, index: Tuple[int, int], value: Union[float, int]) -> None: ...

    def __setitem__(self, index: IndexType, value: Union[float, int]) -> None:
        """
        Set an item while maintaining symmetry.

        Parameters
        ----------
        index : int or tuple of int
            Index or indices where the value should be set.
            Can be either a single integer or a tuple of two integers.
        value : float or int
            The value to set at the specified index/indices.

        Notes
        -----
        When setting element [i,j], element [j,i] is automatically set to
        the same value to maintain symmetry.

        Examples
        --------
        >>> sym = SymmetricArray([[1, 2], [3, 4]])
        >>> sym[0, 1] = 5  # Also sets sym[1, 0] = 5
        """
        # Convert single index to tuple if necessary
        if not isinstance(index, tuple):
            super().__setitem__(index, value)
            return

        i, j = index
        super().__setitem__((i, j), value)
        # Set symmetric element
        if i != j:
            super().__setitem__((j, i), value)
````

#### 88. How to implement the Game of Life using numpy arrays? (★★★)

````python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import HTML
from matplotlib.animation import FuncAnimation

# Enable interactive plotting in Jupyter Notebook
%matplotlib notebook


def update_grid(grid: np.ndarray) -> np.ndarray:
    """
    Update the grid according to the rules of the Game of Life.
    """
    # Count the number of live neighbors for each cell
    neighbors = (
        np.roll(grid, 1, axis=0)
        + np.roll(grid, -1, axis=0)  # Rows above and below
        + np.roll(grid, 1, axis=1)
        + np.roll(grid, -1, axis=1)  # Columns left and right
        + np.roll(np.roll(grid, 1, axis=0), 1, axis=1)  # Diagonals
        + np.roll(np.roll(grid, 1, axis=0), -1, axis=1)
        + np.roll(np.roll(grid, -1, axis=0), 1, axis=1)
        + np.roll(np.roll(grid, -1, axis=0), -1, axis=1)
    )

    # Apply the rules of the Game of Life
    new_grid = np.where(
        (grid == 1) & ((neighbors < 2) | (neighbors > 3)),
        0,
        grid,  # Death
    )
    new_grid = np.where(
        (grid == 0) & (neighbors == 3),
        1,
        new_grid,  # Rebirth
    )

    return new_grid


def animate(frame, img, grid):
    """
    Update the grid and the image for each frame of the animation.
    """
    new_grid = update_grid(grid)
    img.set_data(new_grid)
    grid[:] = new_grid[:]
    return img


# Initialize the grid
grid_size = 50
grid = np.random.choice(
    [0, 1], size=(grid_size, grid_size), p=[0.85, 0.15]
)  # Random initial state

# Set up the animation
fig, ax = plt.subplots()
img = ax.imshow(grid, cmap="binary", interpolation="nearest")
ani = FuncAnimation(
    fig, animate, fargs=(img, grid), frames=100, interval=200, blit=False
)

# Display the animation in the notebook
HTML(ani.to_jshtml())
````

#### 90. Given an arbitrary number of vectors, build the cartesian product (every combinations of every item) (★★★)

Solution from [Stack Overflow](https://stackoverflow.com/a/11146645/12923148):

````python
from typing import List

import numpy as np


def cartesian_product(*arrays: np.ndarray) -> np.ndarray:
    num_arrays: int = len(arrays)
    dtype: np.dtype = np.result_type(*arrays)
    array_lens: List[int] = [len(array) for array in arrays]
    array_lens.append(num_arrays)
    output: np.ndarray = np.empty(array_lens, dtype=dtype)
    for i, array in enumerate(np.ix_(*arrays)):
        output[..., i] = array
    return output.reshape(-1, num_arrays)
````

````python
def find_subset_matches(a: np.ndarray, b: np.ndarray) -> np.ndarray:
    matches = np.zeros((a.shape[0], b.shape[0]), dtype=bool)

    # For each row in a
    for i in range(len(a)):
        # For each row in b
        for j in range(len(b)):
            # Check if all elements in current row of b exist in current row of a
            matches[i, j] = np.all(np.isin(b[j], a[i]))

    # A row in a is a match if it contains elements from all rows in b
    final_matches = np.any(matches, axis=1)

    return final_matches


matching_rows_mask = find_subset_matches(a, b)

print("Original array a:")
print(a)
print("\nArray b to match against:")
print(b)
print("\nRows in a that contain elements from each row of b:")
print(a[matching_rows_mask])
print("\nMatching row indices:")
print(np.where(matching_rows_mask)[0])
````

#### 98. Considering a path described by two vectors (X, Y), how to sample it using equidistant samples (★★★)?

#### 1. Pairwise Distances

We have a sequence of points $\{(x_i, y_i)\}_{i=0}^{n-1}$, where $X$ and $Y$ are arrays of their respective coordinates:
$$
X = (x_0, x_1, \ldots, x_{n-1}), 
\quad 
Y = (y_0, y_1, \ldots, y_{n-1}).
$$

For each consecutive pair $(x_i, y_i)$ and $(x_{i+1}, y_{i+1})$, the Euclidean distance is:

$$
\mathrm{distance}_i 
   = \sqrt{(x_{i+1} - x_i)^2 + (y_{i+1} - y_i)^2}
   \quad \text{for} \quad i = 0, 1, \ldots, n-2.
$$

#### 2. Cumulative Distance

Define the cumulative distance array $CD$ so that $CD[i]$ is the distance along the path from the first point (index 0) up to the $i$-th point. We start at 0 distance for $i=0$. Therefore,

$$
CD[0] = 0,
$$

and for $i \ge 1$,

$$
CD[i] 
   = \sum_{j=0}^{i-1} \mathrm{distance}_j.
$$

Concretely:

$$
CD[1] = \mathrm{distance}_0,\quad
CD[2] = \mathrm{distance}_0 + \mathrm{distance}_1,\quad 
\ldots,\quad
CD[n-1] = \sum_{j=0}^{n-2} \mathrm{distance}_j.
$$

In Python (as shown in the code), one uses:

```python
distances = np.sqrt(np.diff(X)**2 + np.diff(Y)**2)
cumulative_distance = np.cumsum(distances)
cumulative_distance = np.insert(cumulative_distance, 0, 0)
```

which inserts a 0 at the start, thus matching the above formula.

#### 3. Determine Equidistant Sample Points

Let $N$ be the number of desired output (resampled) points. We define a new set of distances

$$
\text{new\_distances} 
  = \{\, d_0, d_1, \ldots, d_{N-1} \},
$$

that are equally spaced from 0 to the total length of the curve, $CD[n-1]$. This is done by

$$
\text{new\_distances} 
  = \mathrm{linspace}\bigl(0,\; CD[n-1],\; N\bigr).
$$

In Python:

```python
new_distances = np.linspace(0, cumulative_distance[-1], N)
```

#### 4. Interpolation to Get the New Points

Finally, we want the coordinates $(X_{\text{new}}, Y_{\text{new}})$ for points that lie at these equidistant “distances along the path.” Since we know how $X$ and $Y$ evolve with respect to $CD$, we can invert that relationship with 1D interpolation:

$$
X_{\text{new}}[k] 
  = \mathrm{interp}\bigl(d_k,\; CD,\; X\bigr),
$$

$$
Y_{\text{new}}[k] 
  = \mathrm{interp}\bigl(d_k,\; CD,\; Y\bigr),
$$

for $k = 0, 1, \ldots, N-1$, where $d_k = \text{new\_distances}[k]$.

In Python:

```python
X_new = np.interp(new_distances, cumulative_distance, X)
Y_new = np.interp(new_distances, cumulative_distance, Y)
```


````python
import numpy as np

# Define the original path as two vectors (X, Y)
X = np.array([0, 1, 2, 3, 4])  # X-coordinates
Y = np.array([0, 1, 0, 1, 0])  # Y-coordinates

# Step 1: Compute pairwise distances
distances = np.sqrt(np.diff(X, axis=-1) ** 2 + np.diff(Y, axis=-1) ** 2)

# Step 2: Calculate cumulative distances
cumulative_distance = np.cumsum(distances)
cumulative_distance = np.insert(cumulative_distance, 0, 0)  # Start at 0

# Step 3: Determine equidistant sample points
N = 10  # Number of equidistant points
new_distances = np.linspace(0, cumulative_distance[-1], N)

# Step 4: Interpolate X and Y coordinates
X_new = np.interp(new_distances, cumulative_distance, X)
Y_new = np.interp(new_distances, cumulative_distance, Y)

# Output the new sampled path
print("Equidistant X:", X_new, "\n")
print("Equidistant Y:", Y_new)
````

#### 99. Given an integer n and a 2D array X, select from X the rows which can be interpreted as draws from a multinomial distribution with n degrees, i.e., the rows which only contain integers and which sum to n. (★★★)

To solve this problem, we need to identify rows in a 2D array `X` that satisfy two conditions:

1. All elements in the row are integers.
2. The sum of the elements in the row equals `n`.

#### Steps

1. **Check for Integers**:
   - Use `np.equal(np.mod(X, 1), 0)` to check if all elements in a row are integers. This works because the modulo operation with 1 will yield 0 for integers.

2. **Check the Sum**:
   - Use `np.sum(row) == n` to check if the sum of the row equals `n`.

3. **Combine Conditions**:
   - Use boolean indexing to filter rows that satisfy both conditions.

````python
import numpy as np


def select_multinomial_rows(X: np.ndarray, n: int) -> np.ndarray:
    # Check if all elements in each row are integers
    is_integer = np.equal(np.mod(X, 1), 0).all(axis=1)

    # Check if the sum of each row equals n
    sums_to_n = np.sum(X, axis=1) == n

    # Combine the two conditions
    valid_rows = X[is_integer & sums_to_n]

    return valid_rows


X = np.array(
    [
        [1, 2, 3],  # Valid: integers and sum to 6
        [1.5, 2.5, 2],  # Invalid: not all integers
        [0, 3, 3],  # Valid: integers and sum to 6
        [2, 2, 2],  # Valid: integers and sum to 6
        [1, 1, 1],  # Invalid: sum is 3, not 6
    ]
)

n = 6  # Desired sum

# Select valid rows
valid_rows = select_multinomial_rows(X, n)
valid_rows
````

#### Explanation

1. **`np.mod(X, 1)`**:
   - Computes the remainder when each element of `X` is divided by 1. For integers, this will always be 0.

2. **`np.equal(np.mod(X, 1), 0)`**:
   - Returns a boolean array where `True` indicates that the corresponding element is an integer.

3. **`.all(axis=1)`**:
   - Ensures that all elements in a row are integers (i.e., all values in the row are `True`).

4. **`np.sum(X, axis=1)`**:
   - Computes the sum of each row.

5. **Boolean Indexing**:
   - Filters rows that satisfy both conditions (`is_integer` and `sums_to_n`).

#### Explanation of Output

- The first row `[1, 2, 3]` is valid because all elements are integers and their sum is 6.
- The second row `[1.5, 2.5, 2]` is invalid because it contains non-integer values.
- The third row `[0, 3, 3]` is valid because all elements are integers and their sum is 6.
- The fourth row `[2, 2, 2]` is valid because all elements are integers and their sum is 6.
- The fifth row `[1, 1, 1]` is invalid because its sum is 3, not 6.

#### Edge Cases

1. **Empty Input**:
   - If `X` is empty, the function will return an empty array.

2. **No Valid Rows**:
   - If no rows satisfy the conditions, the function will return an empty array.

3. **Negative Integers**:
   - The function will work with negative integers as long as they sum to `n`.

4. **Non-integer `n`**:
   - If `n` is not an integer, the function will still work, but it’s unlikely to find valid rows unless `X` contains non-integer values that sum to `n`.

#### Complexity

- **Time Complexity**: $O(n \cdot m)$, where $n$ is the number of rows and $m$ is the number of columns in `X`.
- **Space Complexity**: $O(n \cdot m)$ for storing the result.

#### 100. Compute bootstrapped 95% confidence intervals for the mean of a 1D array X (i.e., resample the elements of an array with replacement N times, compute the mean of each sample, and then compute percentiles over the means). (★★★)

Given the sample `x`, we sample with replacement to obtain a new sample $x_{\text{bootstrap}}$ containing $5,000$ observations (may be duplicated) from this original sample `x`. We repeat this sampling procedure many times--- say, $10,000$--- to obtain many samples:

* $x_{\text{bootstrap 1}}$ ($5,000$ element vector)

* $x_{\text{bootstrap 2}}$ ($5,000$ element vector)

* ...

* $x_{\text{bootstrap 10,000}}$ ($5,000$ element vector)

Then, for each of the bootstrapped samples, we compute the sample mean, generating 10,000 bootstrapped sample means. Finally, we use these bootstrapped statistics to construct the sampling distribution of the sample mean of the original sample `x`, from which the bootstrapped confidence interval can be obtained by taking the appropriate percentiles.

````python
boots_means = []

for i in tqdm(range(10000)):
    mean = (
        np.random.choice(x, size=5000, replace=True).mean()
    )  # With replacement and assumes uniform distribution over all entries of 'x'
    boots_means.append(mean)
````

Plot the sampling distribution and the sample mean of the original array `x`:

````python
fig, ax = plt.subplots(figsize=(15, 12))
sns.histplot(boots_means, kde=True, ax=ax)
# Add the mean of the original distribution
ax.axvline(x.mean(), color="red", linestyle="--")
# Add the text label for the mean of the original distribution
ax.text(
    x.mean() + 10,
    500,
    f"Mean of Original Distribution: {round(x.mean(), 4)}",
    color="red",
    fontsize=12,
)
plt.title("Histogram and Kernel Density Estimate of Bootstrapped Sample Means")
plt.show();
````

We first fit a continuous distribution to the sampling distribution, obtaining MLE estimates of the parameters:

````python
# Bounds are not necessary, but it is good practice to specify them for convergence
# Bound mean to be between the 25th and 75th quantiles values of the distribution
# Bound standard deviation to be positive
quantiles = np.quantile(boots_means, [0.25, 0.75])
bounds = {"loc": (quantiles[0], quantiles[1]), "scale": (0, 1000)}
fit_res = fit(dist=norm, data=boots_means, bounds=bounds, guess=None, method="mle")

fit_res
````

We then test the null hypothesis that the sampling distribution is normally distributed with the MLE estimates. Note that in the context of `scipy.stats.norm`, the `loc` parameter is the mean and the `scale` parameter is the standard deviation:

````python
kstest(boots_means, cdf=norm.cdf, args=(fit_res.params[0], fit_res.params[1]))
````

Controlling type-I error rate at $5 \%$, the probability of observing a test statistic as large as seen above is very large; we fail to reject the null hypothesis that the bootstrapped sampling distribution is normal with the MLE estimators as its parameters. 

In a one-sample test, `statistic_location` is the value of `boots_means` corresponding with the KS statistic--- in other words, the distance between the empirical distribution function and the hypothesized cumulative distribution function is measured *at this observation*.

In addition, for one-sample tests, the `statistic_sign` is $+1$ if the KS statistic is the **maximum positive difference** between the empirical distribution function and the hypothesized cumulative distribution function ($D+$); it is $-1$ if the KS statistic is the **maximum negative difference** ($D-$).

````python
from scipy.stats import t

# Compute t-score (rather than z-score from standard normal)
# In practice, either one can be used for large sample sizes, but we do not know the population standard deviation and so we will use t-score
t_score = t.ppf(q=0.975, df=len(boots_means) - 1)
n = len(boots_means)

# Construct the 95% confidence interval of the bootstrapped sample means using t-distribution
# Use ddof=1 (n - 1) to get the unbiased estimate of the population standard deviation
lower_ci = np.mean(boots_means) - t_score * (np.std(boots_means, ddof=1) / np.sqrt(n))
upper_ci = np.mean(boots_means) + t_score * (np.std(boots_means, ddof=1) / np.sqrt(n))

print(f"95% Confidence Interval: ({round(lower_ci, 4)}, {round(upper_ci, 4)})")
````

What we have calculated above is, strictly speaking, simply the 95 percent confidence interval of the normal distribution. Because the distribution is indeed normal, the confidence interval is very tight.

To compute the actual bootstrapped confidence interval, we need to find the appropriate percentile or quantile values.

````python
boots_confint = np.percentile(boots_means, [2.5, 97.5])

print(
    f"95% Confidence Interval: ({round(boots_confint[0], 4)}, {round(boots_confint[1], 4)})"
)
````

### Confidence Interval for Exponential Distribution

The confidence interval for the sample mean of the exponentially distributed sample has a closed form formula. From the textbooks 'Introduction to Probability and Statistics for Engineers and Scientists' by Sheldon Ross, the $100(1-a) \%$ confidence interval for the rate parameter of an exponential distribution is given by:

$$
\frac{2 n}{\widehat{\lambda} \chi_{1-\frac{\alpha}{2}, 2 n}^2}<\frac{1}{\lambda}<\frac{2 n}{\widehat{\lambda} \chi_{\frac{\alpha}{2}, 2 n}^2}
$$
which is also equal to:
$$
\frac{2 n \bar{x}}{\chi_{1-\frac{\alpha}{2}, 2 n}^2}<\frac{1}{\lambda}<\frac{2 n \bar{x}}{\chi_{\frac{\alpha}{2}, 2 n}^2}
$$

where 

* $\chi_{p, v}^2$ is the $100(p)$ percentile of the chi squared distribution with $v$ degrees of freedom
  * $p$ is $1 - \frac{\alpha}{2}$
  * $v$ is $2n$
* $\mathrm{n}$ is the number of observations of inter-arrival times in the sample
* $\bar{x}$ is the sample average

Note that the `scale` parameter in `numpy.random.exponential` is $\beta$, which is the inverse of the rate parameter $\lambda=\frac{1}{\beta}$. In other words, $\beta=\frac{1}{\lambda}$ and we set this to $1,000$ above.

````python
v = 2 * n  # Degrees of freedom
p_lower = 1 - 0.05 / 2
p_upper = 0.05 / 2
chi_2_percentile_lower = chi2.ppf(q=p_lower, df=v)
chi_2_percentile_upper = chi2.ppf(q=p_upper, df=v)

lower_ci_expon = (2 * n * np.mean(boots_means)) / chi_2_percentile_lower
upper_ci_expon = (2 * n * np.mean(boots_means)) / chi_2_percentile_upper

print(
    f"95% Confidence Interval (Exponential Distribution): ({round(lower_ci_expon, 4)}, {round(upper_ci_expon, 4)})"
)
````
